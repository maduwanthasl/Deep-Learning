**Mean Squared Error**

```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
y_true = [0.0,0.5,0.7,1]
y_pred = [0.2,0.3,0.9,0.7]

y_pred = np.array(y_pred)
y_true = np.array(y_true)

mean_squared_error = tf.keras.losses.MeanSquaredError()
result = mean_squared_error(y_true,y_pred).numpy()
print(f'Mean Squared Error: {result}')

```

```python

# Plotting both y_true and the MSE values
plt.plot(y_true, label='y_true')
plt.plot([result] * len(y_true), label='Mean Squared Error', linestyle='--')
plt.scatter(range(len(y_true)), y_true, color='blue', label='y_true points')
plt.scatter(range(len(y_true)), [result] * len(y_true), color='red', label='MSE value', zorder=5)
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Plot of y_true and Mean Squared Error')
plt.legend()
plt.grid(True)
plt.show()
```

**Mean Absolute Error**

```python
mean_squared_error = tf.keras.losses.MeanAbsoluteError()
result = mean_squared_error(y_true,y_pred).numpy()
print(f'Mean Absolute Error: {result}')
```

```python

# Plotting both y_true and the MSE values
plt.plot(y_true, label='y_true')
plt.plot([result] * len(y_true), label='Mean Absolute Error', linestyle='--')
plt.scatter(range(len(y_true)), y_true, color='blue', label='y_true points')
plt.scatter(range(len(y_true)), [result] * len(y_true), color='red', label='MAE value', zorder=5)
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Plot of y_true and Mean Absolute Error')
plt.legend()
plt.grid(True)
plt.show()
```
