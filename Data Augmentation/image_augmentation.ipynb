```python
import tensorflow as tf
```

```python
from tensorflow.keras.utils import img_to_array, load_img
```

```python
import requests
from keras.preprocessing.image import load_img

# URL of the image
url = "https://st.depositphotos.com/1927453/1975/i/380/depositphotos_19750405-stock-photo-alone-tree-on-meadow-at.jpg"
# Download the image
response = requests.get(url)
image_path = "tree.jpg"  # Save the image with a local filename

# Save the image to a file
with open(image_path, 'wb') as f:
    f.write(response.content)

# Load the image using load_img
img = load_img(image_path)
```

```python
x  = img_to_array(img)
```

```python
x.shape
```

```python
x = x.reshape((1,)+x.shape)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(
    rotation_range = 40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    rescale = 1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True

)
```

```python
import matplotlib.pyplot as plt

i = 0
for batch in datagen.flow(x, batch_size=1, save_to_dir=None):
    plt.figure(i)
    imgplot = plt.imshow(batch[0])
    i += 1
    if i > 10:
        break
plt.show()
```
